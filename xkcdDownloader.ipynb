### A Routine to Download All XKCD Comics With Alt-Text

```python
import os
f = 'xkcd_images'
if not os.path.exists(f):
    os.makedirs(f)
```

```python
xkcdArchiveURL = 'http://xkcd.com/archive/'
from bs4 import BeautifulSoup 
```

```python
import urllib2

xkcdArchiveObj = urllib2.urlopen(xkcdArchiveURL)
xkcdArchivePage = xkcdArchiveObj.read()
print(xkcdArchivePage)
```

```python
soup = BeautifulSoup(xkcdArchivePage)
print(soup.prettify())
```

These numbers are the links before and after the list of comics in the archive. They were determined through manual inspection

```python

nonComicLinksBefore = 7
nonComicLinksAfter = 13
links = soup.find_all('a')[nonComicLinksBefore:-nonComicLinksAfter]
for i in range(len(links)):
    print links[i]
```

```python
hrefs=[]
for link in links:
    print(link.get('href'))
    hrefs.append(link.get('href'))
```

```python
xkcdComicPageObj = urllib2.urlopen("http://xkcd.com"+hrefs[1])
xkcdComicPage = xkcdComicPageObj.read()
soup = BeautifulSoup(xkcdComicPage)
print(soup.prettify())
```

```python
#one way of doing things:
for div in soup.find_all('div', attrs={"id": "comic"}):
    print div.find('img')['src']
    print div.find('img')['title']
    
```

```python
#a simpler approach:
comic = soup.find_all('div', attrs={"id": "comic"})
print(comic[0].find('img')['src'])
```

```python
import Image
import cStringIO
from PIL import ImageFont
from PIL import ImageDraw

file = cStringIO.StringIO(urllib2.urlopen(comic[0].find('img')['src']).read())
im = Image.open(file)
oldImSize = im.size
#break up the alt-text string:
imTextLength = int(oldImSize[0] / 10.)
altText = comic[0].find('img')['title']
altTextLen = len(altText)
numLines = int(altTextLen / imTextLength) + 1
print(numLines)
newAltText = []
for i in range(numLines):
    newAltText.append(altText[i * imTextLength: (i + 1) * imTextLength])
    print(altText[i * imTextLength: (i + 1) * imTextLength])
textHeight = 25

print(type(oldImSize[0]))
print(oldImSize)
newIm = Image.new("RGB", [oldImSize[0], oldImSize[1] + textHeight * numLines], "white")
newIm.paste(im, (0,0))

draw = ImageDraw.Draw(newIm)
font = ImageFont.truetype("/Library/Fonts/Andale Mono.ttf", 16)
for i in range(numLines):
    draw.text((0, oldImSize[1] + i * textHeight),newAltText[i],(0,0,0),font=font)
newIm.show()
```
